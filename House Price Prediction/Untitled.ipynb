```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20, 10)
```

```python
df1 = pd.read_csv('Bengaluru_House_Data.csv')
df1.head()
```

```python
df1.shape
```

```python
df2 = df1.drop(['area_type', 'availability', 'society', 'balcony'], axis='columns')
df2.head()
```

```python
df2.isna().sum()
```

```python
df3 = df2.dropna()
df3.head()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3.head()
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].head()
```

```python
def convert_float_to_num(x):
    nums = x.split('-')
    if len(nums) == 2:
        return (float(nums[0]) + float(nums[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_float_to_num)
df4.head()
```

```python
df4.isna().sum()
```

# Outlier Remover

# We remove data points which are data errors or sometimes they are not data errors but they just represent the extreme variation you know data set so that although they are valid it makes sense to remove them otherwise they can create some issues later on.

# We can remove such data point by some technique. iit can be standard deviation for ex. or you can use DOMAIN KNOWLEDGE (in our ex. knowledge about house pricing and are it cover i.e total_area/bedroom should not be less than a particular value). In company you will get this information by project manager or some one else. Here we wil assume that 1 bedroom are should not be less than 300

```python
df5[df5['total_sqft']/df5['bhk'] < 300].head(10)
```

```python
df5.shape
```

```python
df5[df5['total_sqft']/df5['bhk'] < 300].shape
```

```python
df6 = df5[~(df5['total_sqft']/df5['bhk'] < 300)]
df6.head()
```

```python
df6.shape
```

```python
df6.price_per_sqft.describe()
```
